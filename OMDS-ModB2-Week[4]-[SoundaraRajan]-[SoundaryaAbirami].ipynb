Histogram vs. Boxplot for Drug Prices

Histogram
Pros:
• Visualizes the full distribution of drug prices, showing how frequently different price ranges occur.
• Reveals details like skewness, modality, and the presence of long tails commonly occurring in price data.
Cons:
• Heavily dependent on bin selection—poor choices can obscure important features.
• Doesn’t clearly show summary statistics like medians or quartiles unless annotated.

Boxplot
Pros:
• Concisely summarizes key statistics (median, quartiles, and potential outliers) that directly answer how expensive the drugs tend to be.
• Clearly highlights the 75th percentile and the range of “typical” prices, making it easy to see where the expensive drugs lie relative to the rest.
Cons:
• Doesn’t display the overall frequency distribution or finer details like multiple peaks in the data.
• May hide the underlying shape of the distribution, such as skewness, which can be important in pricing data.

I'd choose boxplot because Its clear depiction of medians, quartiles, and outliers quickly shows where the top prices fall relative to the bulk of the data

Potential Formatting Enhancements:

Annotate Key Values: Label the 75th percentile and median on the boxplot to make those statistics immediately visible.
Logarithmic Scale: If the price data is highly skewed, applying a logarithmic scale (especially on the histogram) can help reveal details in the lower price range while still highlighting the expensive drugs.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset_phishing.csv')

# Quick look at the data and its structure
print(df.head())
print(df.info())
```

```python
print("Columns in dataset:", df.columns.tolist())

# Based on our inspection, we choose these as the important columns:
#url_length: May indicate suspiciously long URLs.
#domain_age: Phishing sites often have very new domains.
#domain_registeration_length: Short registration periods can be a warning sign.

important_cols = ['length_url', 'domain_registration_length', 'domain_age']

# For each important column, create visualizations:
for col in important_cols:
    # Histogram with KDE overlay
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Histogram and KDE of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
    
    # Violin Plot to visualize distribution details
    plt.figure(figsize=(6, 4))
    sns.violinplot(y=df[col], color='lightgreen')
    plt.title(f'Violin Plot of {col}')
    plt.ylabel(col)
    plt.show()
    
    # Swarm Plot grouped by the categorical 'phishing_label'
    if 'status' in df.columns:
        # Using a sample to prevent overcrowding if the dataset is large
        sample_df = df.sample(n=200, random_state=0)
        plt.figure(figsize=(8, 4))
        sns.swarmplot(x='status', y=col, data=sample_df, palette="Set2")
        plt.title(f'Swarm Plot of {col} by Phishing Label')
        plt.xlabel("Phishing Label")
        plt.ylabel(col)
        plt.show()
```

```python

sns.set(style="whitegrid", context="talk")

plt.figure(figsize=(12, 8))
ax = sns.histplot(
    data=df,
    x="length_url",
    hue="status",           # Group data by the 'status' column
    bins=30,                
    multiple="dodge",       
    palette="viridis",      
    edgecolor=".3",         
    linewidth=.5,           
    stat="count",           
    common_norm=False       # Normalize each group separately
)


ax.set_title("Grouped Histogram of URL Length by Status", fontsize=16)
ax.set_xlabel("URL Length", fontsize=14)
ax.set_ylabel("Count", fontsize=14)


plt.legend(title="Status", fontsize=12, title_fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Identify numeric columns in the dataset
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Calculate and print summary statistics for each numeric column
for col in numeric_cols:
    mean_val = df[col].mean()
    median_val = df[col].median()
    mode_series = df[col].mode()
    mode_val = mode_series.iloc[0] if not mode_series.empty else np.nan
    
    print(f"Column: {col}")
    print(f"  Mean   : {mean_val:.2f}")
    print(f"  Median : {median_val:.2f}")
    print(f"  Mode   : {mode_val}")
    print("-----------------------------")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data to mimic phishing and non-phishing proportions
categories = ['Short URL', 'Long URL', 'New Domain', 'Old Domain']
phishing = [60, 80, 90, 20]      # Percentages for phishing URLs
non_phishing = [40, 20, 10, 80]  # Percentages for non-phishing URLs

# Bar width and position
bar_width = 0.5
index = np.arange(len(categories))

# Create stacked bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# First stack (phishing)
ax.bar(index, phishing, bar_width, label='Phishing', color='darkred')

# Second stack (non-phishing), stacked on top of phishing
ax.bar(index, non_phishing, bar_width, bottom=phishing, label='Non-Phishing', color='lightgray')

# Titles and labels
ax.set_xlabel('Feature Category', fontsize=12)
ax.set_ylabel('Percentage (%)', fontsize=12)
ax.set_title('Distribution of URL Features (Phishing vs. Non-Phishing)', fontsize=14)
ax.set_xticks(index)
ax.set_xticklabels(categories, rotation=15)
ax.legend()

# Gridlines for better readability
ax.yaxis.grid(True, linestyle='--', alpha=0.7)

# Final layout adjustments
plt.tight_layout()
plt.show()
```

Stacked Bars: Each bar is divided into two colored segments: one for "Phishing" and one for "Non-Phishing" percentages.
Minimal Colors: Use of contrasting but non-clashing colors (dark red and light gray) to highlight phishing URLs while keeping the chart clean and simple.
Gridlines: Light gridlines on the y-axis help with estimating values without cluttering the chart.
