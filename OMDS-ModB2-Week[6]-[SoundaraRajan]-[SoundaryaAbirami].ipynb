# 2. Weekly graph question

Suppose you wanted to show the nitrate and phosphate level in a water sample.  What are the advantages and disadvantages of showing this as an area plot, as opposed to two separate line graphs?  How would you adjust the graphs shown to improve the presentation?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series_2 =  100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series_2)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

```python
df = pd.DataFrame({'ser1': time_series, 'ser2': time_series_2}, index = datearray)
df.plot.area()
```

Area plots vs Separate line graphs

Area plots are good for showing total combined levels, compact and space efficient, and provides a strong visual impact when the focus is on cumulative amounts rather than individual trends.
But, area plots have disadvantages where it is hard to compare individual chemical trends clearly, because one variable is stacked on top of the other.

Separate Line graphs
They are very clear for comparing trends — we can easily see when nitrate and phosphate increase, decrease, or move together. It is simple and precise. 
But, separate line graphs have disadvantages where it takes more space, and does not visually emphasize the total combined load.

How to improve graphs:
For Area Plot:
Use of clear legends and distinct colors to differentiate nitrate and phosphate, and labelling each area directly.
For Line graphs:
Label lines directly instead of just the legend for clarity, and use of different markers or line styles to distinguish the chemicals.

```python
numeric_cols = df.select_dtypes(include='number')
numeric_cols
```

```python
# Compute correlation matrix
corr_matrix = numeric_cols.corr()
corr_matrix
```

```python
# Unstack to get pairs and sort them
corr_pairs = corr_matrix.abs().unstack().sort_values(ascending=False).drop_duplicates()

# Filter for strong correlations with 0.8 as threshold
high_corr_pairs = corr_pairs[corr_pairs > 0.8]

# Display top correlated pairs
print(high_corr_pairs.head(10))
```

```python
# Pairplot for a subset of features to see the distribution and relationships
features = ['length_url', 'length_hostname', 'nb_dots', 'nb_hyphens', 'web_traffic']
sns.pairplot(df[features + ['status']], hue='status', palette='viridis')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
# Subset of important numeric columns (e.g., top correlated ones)
subset_cols = ['domain_age', 'domain_registration_length', 'web_traffic', 'page_rank']  # Example

# Pairplot to show relationships
sns.pairplot(df[subset_cols])
plt.show()
```

```python


for i, (label1, label2) in enumerate(high_corr_pairs.index[:5], 1):  # Only first 5 pairs
    sns.scatterplot(x=df[label1], y=df[label2])
    plt.title(f'Scatterplot of {label1} vs {label2}')
    plt.show()

```

```python
# Outlier detection: using boxplots for one of the numerical variables
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='status', y='length_url')
plt.title("Boxplot of URL Length by Status")
plt.xlabel("Status")
plt.ylabel("URL Length")
plt.show()
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 86-98 of the Storytelling With Data book as best you can.  (The second half of chapter three).  You do not have to get the exact data values right, just the overall look and feel.

```python


# Simulated monthly sales data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [23, 45, 56, 78, 56, 67, 89, 54, 43, 67, 78, 80]

# Identify the month with the maximum sales
max_sales = max(sales)
highlight_index = sales.index(max_sales)

# Define colors: most bars in a neutral grey, with the peak highlighted in blue.
colors = ['grey'] * len(sales)
colors[highlight_index] = 'blue'

# Create the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(months, sales, color=colors)


ax = plt.gca()
for spine in ['top', 'right', 'left', 'bottom']:
    ax.spines[spine].set_visible(False)
plt.tick_params(left=False, bottom=False)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 1, int(yval),
             ha='center', va='bottom', fontsize=10)

# Add descriptive title and axis labels
plt.title('Monthly Sales with Highlighted Peak', fontsize=14, weight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales', fontsize=12)

plt.show()
```
